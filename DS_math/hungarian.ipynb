```python
#функция для расчета похожести (или непохожести) двух коробок
def calculate_similarity(box1, box2):
    square1 = abs(box1[0] - box1[1]) * abs(box1[2] - box1[3])
    square2 = abs(box2[0] - box2[1]) * abs(box2[2] - box2[3])
    return abs(square1 - square2)
```
