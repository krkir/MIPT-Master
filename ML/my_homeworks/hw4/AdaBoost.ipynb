```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.metrics import recall_score, accuracy_score
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

### Подготовка данных

```python
chips = pd.read_csv("chips.csv")
geyser = pd.read_csv("geyser.csv")

chips_feat = chips.drop("class", axis=1).values
chips_label = chips["class"]
chips_label = (chips_label == "N").astype(int)*2 - 1

geyser_feat = geyser.drop("class", axis=1).values
geyser_label = geyser["class"]
geyser_label = (geyser_label == "N").astype(int)

chips_feat_train, chips_feat_test, chips_label_train, chips_label_test = train_test_split(chips_feat, chips_label, train_size=0.8)
geyser_feat_train, geyser_feat_test, geyser_label_train, geyser_label_test = train_test_split(chips_feat, chips_label, train_size=0.8)
```

Немного подтюним гиперпараметры для бустинга и для дерева

```python
ada_model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=3))
ada_grid = GridSearchCV(ada_model, {"learning_rate": [1e-3, 1e-2, 1e-1, 1, 10]}, verbose=1)
ada_grid.fit(chips_feat, chips_label)
ada_score = accuracy_score(chips_label, ada_grid.predict(chips_feat))
print(f"for ada: {ada_grid.best_params_}, accuracy: {ada_score}")
```

```python
tree_params = {'criterion': ["gini", "entropy"], 'splitter': ["best", "random"]}
tree = DecisionTreeClassifier(random_state=0)
tree_grid = GridSearchCV(tree, tree_params, n_jobs=-1, verbose=True, cv=4)
tree_grid.fit(chips_feat, chips_label)
tree_score = accuracy_score(tree_grid.predict(chips_feat), chips_label)
print(f"for tree: {tree_grid.best_params_}, accuracy: {tree_score}")
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
           levels=0, alpha=0.5,
           linestyles='-')
```

```python
dt = DecisionTreeClassifier(criterion='entropy').fit(chips_feat_train, chips_label_train)

# AdaBoost
ada = AdaBoostClassifier(DecisionTreeClassifier(max_depth=3), learning_rate=0.1)
```

```python
plt.figure(figsize=(20, 30))
sizes = [2, 3, 5, 8, 13, 21, 34, 55]
accuracy_score_history = []
for i, s in enumerate(sizes):
    ada.n_estimators = s
    m = ada.fit(chips_feat, chips_label)
    accuracy_score_history.append(accuracy_score(chips_label, ada.predict(chips_feat)))
    plt.subplot(4, 2, i+1)
    plt.scatter(chips_feat[:, 0], chips_feat[:, 1], c=chips_label, cmap='autumn')
    plot_svc_decision_function(m)
    plt.title('{} trees'.format(s))
```

```python
plt.plot(sizes, accuracy_score_history)
plt.legend("accuracy")
plt.xlabel = "size"
plt.ylabel = "accuracy"
plt.show()
```

```python
max_ = max(accuracy_score_history)
ind = accuracy_score_history.index(max_)
print(f"best n_estimators = {sizes[ind]}, best accuracy = {max_}")
```
